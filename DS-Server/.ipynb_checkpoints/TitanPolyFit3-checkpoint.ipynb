## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define empirically collected data

```python
# Limelight empirical data

# Limelight camera angle - 45 degrees?
# Floor camera angle     - 47 degrees
# Forward camera angle   - 20 degrees (not used for calculations)

Ty = [20.19, 18.7, 17.71, 16.84, 15.89, 15.0, 7.28, 5.41, -1.08, -5.7, -7.5]  # Independent variable during empirical collection.
limelight_d_measured = [65., 56., 48., 44., 36., 31., 19., 17., 10., 6., 2.5]
Tx_measured = [-3.01, -2.83, -1.37, -2.94, -2.54, -4.9, -2.37, -2.2, -3.6, -4.91, -5.8]
```

```python
# Floor Tape Camera empirical data
#(Assume is converted into percent of axis, origin at upper left of image)
#

tape_d_measured = [56., 44., 31.]
Fy = [84, 129, 194]
Fx_measured = [103, 103, 89]
```

## Compute coefficients for functions to use in formulas on the RoboRIO

### Distance formula coefficients:  *D*(Ty) = $c_1$Ty^2 + $c_2$Ty + $c_3$ 

Compute coefficients for polynomial converting Limelight Ty values into robot distance from retro-reflective tape.  The RoboRIO knowing its distance can be useful, e.g. plan kinematic motion to approach destination positions.  

```python
coefficients = np.polyfit(Ty, limelight_d_measured, 3)
polynomial_function = np.poly1d(coefficients)
xp = np.linspace(-10, 20, 100)
plt.plot(Ty, limelight_d_measured, '.', xp, polynomial_function(xp), '-')
plt.xlabel("$T_y$", fontsize=18)
plt.ylabel("Distance Predicted", fontsize=18)
plt.show()
print("Distance Coefficients with highest order first, last is Ty^0")
print(coefficients)
```

### Predicted Tx formula coefficients:  *Tx*(Ty) = $c_1$Ty^2 + $c_2$Ty + $c_3$

Compute coefficients for polynomial converting measured Limelight Ty values into corresponding Limelight Tx values, regarding the floor tape.  This formula is to model the parallax effects, to detect when Tx is a match to the Ty, and thus indicates the robot is aligned properly to approach a port for hatch/cargo feth/deposit.

```python
coefficients = np.polyfit(Ty, Tx_measured, 2)
polynomial_function = np.poly1d(coefficients)
xp = np.linspace(-10, 20, 100)
plt.plot(Ty, Tx_measured, '.', xp, polynomial_function(xp), '-')
plt.xlabel("$T_y$", fontsize=18)
plt.ylabel("Tx Predicted", fontsize=18)
plt.show()
print("Tx_Predicted Coefficients with highest order first, last is Ty^0")
print(coefficients)
```

### Coresponding Tape Horizontal % FOV as function of Tape Vertical % FOV:

### Fx(Fy) = $c_1$Fy^2 + $c_2$Fy + $c_3$

Compute coefficients for polynomial converting measured USB Camera Fy values into corresponding USB Camera Fx values, regarding the floor tape.  This formula is to model the parallax effects, to detect when Fx is a match to the Fy, and thus indicates the robot is aligned properly to approach a port for hatch/cargo fetch/deposit.

```python
#frame resolution was 320x240
print(Fy)
for x in range(len(Fy)):
    Fy[x] *= (100 / 240)
print(Fy)
coefficients = np.polyfit(Fy, Fx_measured, 2)
polynomial_function = np.poly1d(coefficients)
xp = np.linspace(30, 100, 100)
plt.plot(Fy, Fx_measured, '.', xp, polynomial_function(xp), '-')
plt.xlabel("$F_y$", fontsize=18)
plt.ylabel("Fx Predicted", fontsize=18)
plt.show()
print("Fx_Predicted Coefficients with highest order first, last is Fy^0")
print(coefficients)
```
